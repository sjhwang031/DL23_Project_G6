# üìù3. Vocabulary test

**Studying and learning difficult vocabulary in the text**

```python
#@markdown Vocabilary test - Context 1
import random

vocab_dict = {
    "verbally": "in spoken words and not in writing or actions",
    "mean": "with the help of something",
    "evolutionary": "connected with evolution; connected with slow steady development and change",
    "divide": "to separate into parts; to make something separate into parts",
    "indicate": "to show that something is true or exists",
    "awareness": "knowing something; knowing that something exists and is important",
    "ignorance": "a lack of knowledge or information about something",
    "brag": "to talk too proudly about something you own or something you have done",
    "constantly": "all the time; repeatedly",
    "vast": "extremely large in area, size, amount, etc.",
    "spade": " a garden tool with a broad metal part and a long handle, used for digging"
}

def quiz(num_questions):
    score = 0
    questions = random.sample(list(vocab_dict.items()), num_questions)
    for word, definition in questions:
        print("What is the word for this definition? '{}'".format(definition))
        answer = input("Answer: ")
        if answer.lower() == word.lower():
            print("Correct!")
            score += 1
        else:
            print("Incorrect. The correct answer is '{}'.".format(word))
        print()
    print("Quiz complete. You scored {}/{} points.".format(score, num_questions))

if __name__ == '__main__':
    num_questions = int(input("How many questions would you like to play? "))
    quiz(num_questions)
```
